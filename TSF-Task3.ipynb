#### Importing all the required libraries

```python
import warnings 
warnings.filterwarnings('ignore')
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Reading the Data

```python
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

### Checking Null Values

```python
df.isnull().sum()
```

### Data Processing

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
df = df.drop(['Postal Code'],axis=1)
```

```python
df.head()
```

### Exploratory Data Analysis

```python
sns.pairplot(df)
```

### Correlation Analysis

```python
df.corr()
sns.heatmap(df.corr(), annot=True)
plt.show()
```

### Visualizing Ship Modes

```python
df['Ship Mode'].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.title('SHIP MODES')
plt.pie(df['Ship Mode'].value_counts(), labels=df['Ship Mode'].value_counts().index, autopct='%1.1f%%')
plt.show()
sns.countplot(x=df['Ship Mode'])
```

### Visualizing Segment

```python
df['Segment'].value_counts()
```

```python
sns.pairplot(df,hue="Segment")
```

### Visualizing Category

```python
df['Category'].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.title('Category')
plt.pie(df['Category'].value_counts(), labels = df['Category'].value_counts().index, autopct='%1.1f%%')
plt.show()
sns.countplot(x='Category',data=df,palette='tab10')
```

```python
plt.figure(figsize = (8,8))
plt.title('Sub-Category')
plt.pie(df['Sub-Category'].value_counts(), labels=df['Sub-Category'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

### State-wise Profit

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(15,15))
stpr = df.groupby(['State'])['Profit'].sum().nlargest(50)
stpr.plot.barh()
```

   #### Graph displays that California and New York have the highest profits while Texas and Ohio have the least profits

### Region wise Profit

```python
plt.figure(figsize= (6,6))
plt.title('Region wise Profits')
plt.pie(df['Region'].value_counts(), labels=df['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

### Sales, Profit & Discount Interdependency

```python
plt.style.use('seaborn')
df.plot(kind="scatter",figsize=(12,6), x="Sales",y="Profit",c="Discount",s=20,fontsize=12,colormap='plasma')
plt.ylabel('Profits')
plt.show()
```

#### More the Discount More the Sales but Lesser the Profits

### Profit vs Discount

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

### Profit/Loss and Sales of Each State

```python
pls = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
pls[:].plot.bar(color=['red','black'],figsize=(20,10))
plt.title('Profit/Loss and Sales across the States')
plt.xlabel('States')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```
