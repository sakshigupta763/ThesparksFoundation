### From the given iris dataset, predict the optimum number of cluster and represent it visually

#### Importing the required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
```

#### Reading the Data

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head()
```

```python
x = iris_df.iloc[:,:].values
```

#### Predicting the value of K for the model

```python
k_range = range(1,11)
sse = [] 


for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(iris_df)
    sum_of_squares = km.inertia_
    sse.append(sum_of_squares)
    
print("Sum of Square Value from K = 1 to K = 11 are as followed: ")
sse
```

#### Plotting the results

```python
plt.title('The Elbow Method for predicting the values of K')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of Squares of error (SSE) ') # Within cluster sum of squares
plt.annotate("Elbow is obtained at K = 3", xytext=(4,200),xy=(3,78.8514),arrowprops={'facecolor':'red'})
plt.plot(k_range,sse)
plt.show()
```

##### Thus the value of K here is 3. So the optimum no of cluster for this model is 3
